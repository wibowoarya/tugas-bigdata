### **Inisialisasi SparkSession**

Setelah instalasi, kita perlu menginisialisasi `findspark` dan membuat `SparkSession`. `SparkSession` adalah *entry point* untuk menggunakan fungsionalitas Spark dengan *Dataset* dan *DataFrame* API.

```python
import findspark
from pyspark.sql import SparkSession

findspark.init()

# Create a SparkSession
spark = SparkSession.builder.appName("EDA_with_Spark").getOrCreate()

print("SparkSession created successfully!")
```

### **Verifikasi SparkSession**

Untuk memastikan bahwa SparkSession telah berhasil diinisialisasi, kita bisa menampilkan sesi Spark.

```python
spark
```

### **Memuat Data ke dalam Spark DataFrame**

Sekarang setelah `SparkSession` aktif, kita bisa memuat data ke dalamnya. Umumnya, Anda dapat memuat file dari Google Drive, lokal Colab, atau sumber eksternal lainnya. Berikut adalah contoh memuat file CSV. Jika Anda memiliki tipe file lain (JSON, Parquet), Anda bisa mengganti `.csv` dengan `.json` atau `.parquet` dan mengatur opsi yang sesuai.

```python
# Contoh memuat file CSV
# Pastikan Anda memiliki file CSV di lingkungan Colab Anda atau di Google Drive yang sudah di-mount.
# Ganti 'nama_file_anda.csv' dengan path ke file CSV Anda.

# Untuk file CSV di folder 'sample_data' Colab:
# df = spark.read.csv('/content/sample_data/california_housing_train.csv', header=True, inferSchema=True)

# Untuk file CSV yang diunggah ke sesi Colab atau dari Google Drive:
# Misalnya, jika Anda mengunggah 'my_data.csv' ke Colab:
# df = spark.read.csv('my_data.csv', header=True, inferSchema=True)

# Contoh: Menggunakan salah satu file CSV dari sample_data
data_path = '/content/sample_data/california_housing_train.csv'
df = spark.read.csv(data_path, header=True, inferSchema=True)

print(f"Data berhasil dimuat dari: {data_path}")
print(f"Jumlah baris dalam DataFrame: {df.count()}")
```

### **Melihat Schema dan Sampel Data**

Setelah data dimuat, penting untuk memeriksa skema (struktur kolom dan tipe datanya) dan melihat beberapa baris pertama data untuk memverifikasi bahwa semuanya telah dimuat dengan benar.

```python
print("Skema DataFrame:")
df.printSchema()

print("\nBeberapa baris pertama data:")
df.show(5)
```
